```python
import pandas as pd
df = pd.read_csv(r"C:\Users\moham\OneDrive\Desktop\WebScraping\legacy_app\output\all\all_products_coarsetag_clean.csv")
df['Canonical Text'] = df['Product'] + df['Description']
df
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd

# Embedding
model = SentenceTransformer('all-MiniLM-L6-v2')

text_descriptions = list(df["Canonical Text"])
product_embeddings = model.encode(text_descriptions, normalize_embeddings=True)

# 3. Compute the Matrix
# This compares the 'product_embeddings' against themselves
matrix = util.cos_sim(product_embeddings, product_embeddings)
# Convert to a readable numpy array
sim_matrix = matrix.numpy()
sim_matrix
```

```python
import numpy as np

products_similar = [] # Initialize outside the loop

for i in range(765):
    # Get indices of the top 7 similar items
    indices = np.argsort(sim_matrix[i][-7:])
    
    # Create the dictionary for this specific row
    dic = {int(idx): df.iloc[idx]["Product"] for idx in indices}
    
    # Append this dictionary to our master list
    products_similar.append(dic)

print(products_similar)
```

```python
import networkx as nx
import numpy as np

cluster_list = []
# 2. Set your "Agile Threshold"
threshold = 0.7

# 3. Create a Graph
G = nx.Graph()

# 4. Add edges only if similarity is high enough
rows, cols = np.where(sim_matrix > threshold)
for i, j in zip(rows, cols):
    if i != j: # Don't connect a product to itself
        G.add_edge(i, j)

# 5. Find the clusters (Connected Components)
clusters = list(nx.connected_components(G))

for i, cluster in enumerate(clusters):
    c = [df.iloc[idx]["Product"] for idx in cluster]
    cluster_list.append(c)
print(len(cluster_list))
```

```python
import pandas as pd

# Initialize the column with a default value (e.g., -1 or NaN)
df['Cluster #'] = -1

for i, c in enumerate(cluster_list):
    # Create a mask for rows where 'Product' is in the current cluster
    mask = df['Product'].isin(c)
    
    # Only update the 'Cluster #' for those specific rows
    df.loc[mask, 'Cluster #'] = i+1
# filtered_df = df[df['Cluster #'] == -1]
df.to_csv(r'C:\Users\moham\OneDrive\Desktop\WebScraping\legacy_app\output\all\all_products_subcategories.csv', index=False)
```
